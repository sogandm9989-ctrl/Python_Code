```python
weight = float(input("enter your weight in kilogram:"))
height = float(input("enter your height in meter:"))
bmi = weight / (height ** 2)
print("\f your bmi index is equal to:")
if bmi < 18.5:
    print("underweight")
elif 18.5 <= bmi < 25:
    print("normal")
elif 25 <= bmi < 30:
    print("overweight")
elif 30 <= bmi < 35:
    print("obese")
else:
    print("extremely obese")
```
