```python
import random
roll = random.randint(1, 6)
print("first number:", roll)
if roll == 6:
    print("it's 6, rerolling the dice")
    roll = random.randint(1, 6)
    print("new number:", roll)
if roll == 6:
    print("it's 6 again, rerolling the dice")
    roll = random.randint(1, 6)
    print("new number:", roll)
```
